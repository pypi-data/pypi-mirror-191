```python
import papermill as pm

pm.inspect_notebook('parameter test.ipynb')
```

```python
import json
import pprint

code = ""

f = open("parameter test.ipynb")
notebook_string = f.read()
notebook = json.loads(notebook_string)
# get kernel
kernel = notebook["metadata"]["kernelspec"]["name"]
# get parameters cell
for cell in notebook["cells"]:
    if cell["cell_type"] == "code":
        if "metadata" in cell and "tags" in cell["metadata"]:
            if "parameters" in cell["metadata"]["tags"]:
                code = "".join(cell["source"])

print(notebook_string)
print(kernel[len("conda-"):])
print(code)
                
                
def check_variables(_nb_cron_code):
    exec(_nb_cron_code)
    var = locals()
    var.pop("_nb_cron_code")
    return json.dumps(var)

variables = check_variables(code)
print(variables)

# print(json.dumps(notebook, indent=4, sort_keys=True))
```
